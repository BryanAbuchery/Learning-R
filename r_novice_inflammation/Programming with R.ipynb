```r
fahrenheit_to_celsius <- function(temp_F) {
  temp_C <- (temp_F - 32) * 5 / 9
  return(temp_C)}
```

```r
fahrenheit_to_celsius(32)
```

```r
fahrenheit_to_celsius(212)
```

```r
celsius_to_kelvin <- function(temp_C) {
  temp_K <- temp_C + 273.15
  return(temp_K)}

# freezing point of water in Kelvin
celsius_to_kelvin(0)
```

```r
fahrenheit_to_kelvin <- function(temp_F) {
  temp_C <- fahrenheit_to_celsius(temp_F)
  temp_K <- celsius_to_kelvin(temp_C)
  return(temp_K)
}

# freezing point of water in Kelvin
fahrenheit_to_kelvin(32.0)
```

```r
dry_principle <- c("Don't", "repeat", "yourself", "or", "others")
edges <- function(dry_principle) {
    a <- dry_principle[1]
    b <- dry_principle[length(dry_principle)]
    c <- c(a, b)
    return(c)
}
cat(edges(dry_principle))
```

```r
data <- read.csv(file = "inflammation-01.csv", header = FALSE)
analyze <- function(data) {
    # function that plots our data for mean of inflammation across the 40 days,
    # plots the maximum inflammation temperature against all the 40 days,
    # plots the minimum inflammation temperature against all the 40 days
   
    # find the mean, max and min of temperature against the 40 days
    avg_patient_inflammation <- apply(data, 2, mean)
    max_day_inflammation <- apply(data, 2, max)
    min_day_inflammation <- apply(data, 2, min)
    
    # plot the mean, max and min of temperature against the 40 days
    a <- plot(avg_patient_inflammation)
    b <- plot(max_day_inflammation)
    c <- plot(min_day_inflammation)
    return(a)
    return(b)
    return(c)
}
analyze(data)
```

```r
center <- function(data, midpoint = 0) {
  # return a new vector containing the original data centered around the
  # midpoint (0 by default).
  # Example: center(c(1, 2, 3), 0) => c(-1, 0, 1)
  new_data <- (data - mean(data)) + midpoint
  return(new_data)
}
```
